# AULA 2 - LEYENDO ARCHIVOS DE EXCEL

## Importamos archivos de excel

```python
import pandas as pd
# importamos pandas
```

```python
archivo = 'emisiones_CO2.xlsx'
# guardamos el nombre del archivo en una variable
```

```python
datos_excel = pd.read_excel(archivo)
datos_excel.head()
```

```python
# para examinar cuantas hojas tengo en el excel se obtiene esto
pd.ExcelFile(archivo).sheet_names
```

## Parametros de la funcion read_excel
 documentacion :

```python
# la hoja que se abre por defecto es la primera
# para poder selecionar la hoja se agrega el nombre de la hoja que se queirea abrir
percapita = pd.read_excel(archivo, sheet_name='emisiones_percapita')
percapita.sample(5)
# muestra registros alteatorios
```

```python
# la hoja que se abre por defecto es la primera
# para poder selecionar la hoja se agrega el nombre de la hoja que se queirea abrir
fuentes = pd.read_excel(archivo, sheet_name='fuentes')
fuentes.tail()
# muestra registros alteatorios
```

```python
# use cols seleciona las columnas entre un intervalo desde la a la d
intervalo = pd.read_excel(archivo, sheet_name='emisiones_C02',usecols='A:D')
intervalo
```

```python
# use nrows para seleccioanr un intervalo de filas, cantidad definida
intervalo_2= pd.read_excel(archivo, sheet_name='emisiones_C02',usecols='A:D', nrows=10)
intervalo_2
```

# **EXPORTAR DATOS EXCEL**

```python
intervalo.to_excel('c20_percapita.xlsx', index = False)
# siempre con "to" + mas el tipo de dato que queremos exportar
# siempre que exportamos no llevar el indice
```

```python
pd.read_excel('c20_percapita.xlsx')
#leemo el excel
```
